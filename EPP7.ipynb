Crie um programa que gere um arquivo "exemplo.txt". Escreva "n" valores inteiros aleatórios na faixa entre 0~10 e feche o arquivo. Reabra o arquivo gerado, calcule a média dos valores lidos e feche novamente o arquivo.

Obs: Para gerar valores aleatoriamente utilizamos o procedimento rand() da biblioteca stdlib.h. Esse procedimento retorna um valor pseudo-aleatório.

Por exemplo:

Se usarmos "int x = rand()%10;", então x receberá um valor aleatório entre 0 e 9.

Se o programa for executado múltiplas vezes, a sequência de valores aleatórios será sempre igual. Para controlar a geração de valores, é preciso definir uma semente usando o procedimento "srand(seed)".

 Faça com que o valor de seed seja forneceido pelo usuário, e então gere a matriz aleatória. Observe que com valores diferentes de seed, as matrizes em diferentes execuções terão diverentes valores.

Sobre o programa, pressupõe-se a seguinte estrutura: (i) abertura, escrita de dados e fechamento do arquivo; (ii) abertura, leitura dos dados e fechamento do arquivo.




Exemplo de execução:
Digite a quantidade de números:                                                 
5                                                                               
Digite o valor da seed:                                                         
0                                                                               
6 10 6 2 1                                                                      
Média = 5.00  

```python
%%writefile arquivoSimples.c
#include <stdio.h>
#include <stdlib.h>

int main() {
    // Declaração de variáveis
    int n, seed;
    FILE *file;

    // Entrada do usuário
    printf("Digite a quantidade de números: ");
    scanf("%d", &n);
    printf("Digite o valor da seed: ");
    scanf("%d", &seed);

    // Define a seed para a geração de números aleatórios
    srand(seed);

    // Abertura do arquivo para escrita
    file = fopen("exemplo.txt", "w");
    if (file == NULL) {
        printf("Erro ao abrir o arquivo para escrita.\n");
        return 1;
    }

    // Geração e escrita de valores aleatórios no arquivo
    int valores[n];
    for (int i = 0; i < n; i++) {
        valores[i] = rand() % 11; // Gera número aleatório entre 0 e 10
        fprintf(file, "%d ", valores[i]);
    }
    fclose(file);

    // Exibe os números gerados na mesma linha, simulando o comportamento esperado
    for (int i = 0; i < n; i++) {
        printf("%d ", valores[i]);
    }
    printf("\n");

    // Reabre o arquivo para leitura
    file = fopen("exemplo.txt", "r");
    if (file == NULL) {
        printf("Erro ao abrir o arquivo para leitura.\n");
        return 1;
    }

    // Leitura dos números e cálculo da média
    int soma = 0, count = 0, num;
    while (fscanf(file, "%d", &num)!=EOF) {
        soma += num;
        count++;
    }
    fclose(file);

    // Calcula e exibe a média

        double media = (double)soma / n;
        printf("Média = %.2f\n", media);

    return 0;
}
```
