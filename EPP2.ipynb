# Crie um programa que retorne o fatorial  de um número inteiro fornecido pelo usuário usando a instrução “while”.
Exemplo de execução:
Digite um número inteiro:
10
10! = 3628800

```python
%%writefile repeticao.c
#include <stdio.h>
int n;

int fatorial(int n){
  if(n==0|| n==1){
    return 1;
  } else {
    return n*fatorial(n-1);// usando recursão
  }
}

 int main(){
  printf("Digite um número inteiro: ");
  scanf("%d", &n);
  printf("%d! = %d", n, fatorial(n));
  return 0;
 }
```

```python
!gcc repeticao.c -o repeticao
```

```python
!./repeticao
```

Crie um programa que use a instrução “for” para identificar os divisores de um número inteiro. O usuário deverá fornecer um valor inteiro e o programa deverá retornar a mensagem “O número X é divisível por Y” para cada Y divisor de X. Se X for primo (divisível apenas por 1 e por ele mesmo), além dos divisores o programa deverá exibir a mensagem: "O número X é primo.".


Exemplos de execução:
Digite um número inteiro:
10
O número 10 é divisível por  1.
O número 10 é divisível por  2.
O número 10 é divisível por  5.
O número 10 é divisível por  10.


Digite um número inteiro:
7
O número 7 é divisível por  1.
O número 7 é divisível por  7.
O número 7 é primo.

```python
%%writefile repeticao2.c
#include <stdio.h>

int n,div;
int divisores(int n){
  div = 0;
  for(int i = 1; i<=n;i++){
    if(n%i==0 ){
      printf("O número %d é divisível por %d\n", n, i);
      div++;
    }
  }
  if(div==2){
    printf("O número %d é primo\n", n);
  }
}
int main(){
  printf("Digite um número inteiro: ");
  scanf("%d", &n);
  divisores(n);
  return 0;
}
```

```python
!gcc repeticao2.c -o repeticao2
```

```python
!./repeticao2
```
