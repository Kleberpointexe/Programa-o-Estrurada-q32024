Crie um programa que verifica se um número digitado pelo usuário é primo. Este programa deve ser baseado em uma função recursiva decideSePrimo(). Esta função recebe como parâmetros 2 valores: o número testado e um divisor. Então, retorna 1 se for primo ou 0 caso contrário. A função recursiva considera 3 situações:
(i) se divisor ==1, retorna 1
(ii) se número%divisor==0, retorna 0
(iii)caso contrário, retorna decideSePrimo(número, divisor-1)



Utilize como declaração da função:
int decideSePrimo(int num, int div){
     //Código da Função
}


Exemplo de execução:
Digite um valor inteiro: 229
O número 229 é primo.

```python
%%writefile recursao2.c
#include <stdio.h>
int n,d;
int decideSePrimo(int n,int d){
  if(d==1){
    return 1;
  } else if(n%d==0){
    return 0;
  } else {
    return decideSePrimo(n,d-1);
  }
}

int main(){
  printf("Digite um valor inteiro:\n");
  scanf("%d",&n);
  d=n-1;
  if(decideSePrimo(n,d)==1){
    printf("O número %d é primo.",n);
  } else {
    printf("O número %d não é primo.",n);
  }
  return 0;
}
```

Crie um programa que leia 5 números inteiros positivos (utilize uma função inteiroPositivo que leia um número e verifique se ele é positivo, retornando-o caso afirmativo). Para cada número inteiro positivo, escrever a soma de seus divisores(exceto ele mesmo).Utilize a função SomaDivisores para obter a soma.
Nome: somaDivisores
Descrição: Calcula a soma dos divisores do número informado(exceto ele mesmo).
Entrada: Um número inteiro e positivo.
Saída: A soma dos divisores.

```python
%%writefile recursao3.c
#include <stdio.h>

int numbers[5], i;

// Função para garantir que o número seja positivo
int inteiroPositivo(int number) {
    if (number > 0) {
        return number; // Se o número for positivo, retorna-o
    } else {
        printf("Número inválido, insira um número positivo: ");
        scanf("%d", &number); // Solicitar o número novamente se for negativo
        return inteiroPositivo(number); // Chamada recursiva para garantir número positivo
    }
}

// Função para calcular a soma dos divisores de todos os números
int somaDivisores() {
    int soma = 0; // Reinicializar a soma
    for (i = 0; i < 5; i++) {
        for (int j = 1; j < numbers[i]; j++) {
            if (numbers[i] % j == 0) {
                soma += j;
            }
        }
    }
    return soma;
}

int main() {
    // Solicitar 5 números inteiros positivos
    for (i = 0; i < 5; i++) {
        printf("Digite um número inteiro positivo:\n");
        scanf("%d", &numbers[i]);
        numbers[i] = inteiroPositivo(numbers[i]); // Garantir que o número seja positivo
    }

    // Calcular a soma dos divisores de todos os números
    int resultado = somaDivisores();
    printf("A soma de todos os divisores é: %d\n", resultado);

    return 0;
}
```
