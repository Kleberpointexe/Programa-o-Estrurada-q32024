Escreva um programa que exiba o conteúdo de um vetor de inteiros digitado pelo usuário. A entrada dos valores e a exibição devem ser feitas usando duas funções separadas.
Exemplo:
read_vec(vetor, N); // função armazena valores em vetor[]
show_vec(vetor,N); // função exibe valores de vetor[]

Exemplo de execução:


Digite o tamanho do vetor:                                                      
3                                                                               
Digite os elementos do vetor:                                                   
3                                                                               
2                                                                               
1                                                                               
O vetor é:                                                                      
3 2 1

```python
%%writefile vetoresEfuncoes.c
#include<stdio.h>
int N;

void read_vec(int vetor[],int N){
  printf("Digite os elementos do vetor");
  for(int i=0; i<N;i++){
    scanf("%d",&vetor[i]);
  }
  }


int show_vec(int vetor[],int N){
  printf("O vetor é:");
  for(int i=0; i<N; i++){
    printf("%d",vetor[i]);
  }
}

int main(){
  printf("Digite o tamanho do vetor: ");
  scanf("%d",&N); // atribuir o tamanho do vetor
  int vetor[N]; // criar o vetor com tamanho atribuido
  read_vec(vetor, N); // função armazena valores em vetor[]
  show_vec(vetor,N); // função exibe valores de vetor[]
  return 0;
}
```

Criar um programa que eleve todos os elementos de um vetor ao quadrado, salvando os valores no próprio vetor. Use as funções do exercício anterior como base para entrada dos valores e exibição do vetor antes e depois da operação.

Usar as funções: square_vec(vetor, N); show_vec(vetor, N); read_vec(vetor, N);

Exemplo de execução:
Digite o tamanho do vetor:                                                      
2                                                                               
Digite os elementos do vetor:                                                   
23                                                                              
27                                                                              
                                                                                
O vetor é: 23 27                                                                
O vetor é: 529 729

```python
%%writefile vetoresEfuncoes2.c
#include <stdio.h>
#include <math.h>

void read_vec(int vec[],int t){//função para ler o vetor
  printf("Digite os elementos do vetor:");
  for(int i=0;i<t; i++){
    scanf("%d",&vec[i]);
  }
}

void show_vec(int vec[],int t){//função para mostrar o vetor
  printf("O vetor é:");
  for(int i=0;i<t;i++){
    printf("%d\n",vec[i]);
  }
}

void square_vec(int vec[],int t){
  for(int i=0; i<t;i++){
    vec[i]=vec[i] * vec[i];
  }
}

int main(){
  int tamanho;
  printf("Digite o tamanho do vetor:\n");
  scanf("%d",&tamanho);
  int vetor[tamanho];
  read_vec(vetor, tamanho);
  square_vec(vetor,tamanho);
  show_vec(vetor,tamanho);
  return 0;
}
```

Crie um programa que peça ao usuário as dimensões de uma matriz e depois leia cada um de seus elementos. O programa deve então exibir a matriz usando uma função chamada exibeMatriz(). Esta função deve ter como parâmetro as dimensões da matriz, bem como a própria matriz.


Utilize como declaração da função:
void exibeMatriz(int a, int b, int m[a][b]){
     //Código da Função
}



Exemplo de execução:
Digite o tamanho da matriz: 3
2
Digite o elemento (0,0): 1
Digite o elemento (0,1): 2
Digite o elemento (1,0): 3
Digite o elemento (1,1): 4
Digite o elemento (2,0): 5
Digite o elemento (2,1): 6
A matriz digitada é:
1 2

5 6

```python
%%writefile  matrizesEfuncoes3.c
#include <stdio.h>


void read_matriz(int row, int col,int m[][col]){
  for(int i=0;i<row;i++){
    for(int j = 0; j<col;j++){
      printf("Digite o elemento (%d,%d): \n",i,j);
      scanf("%d",&m[i][j]);
    }
  }
}

void exibe_matriz(int row, int col, int m[][col]){
  printf("A matriz digitada é: \n");
  for(int i=0;i<row;i++){
    for(int j = 0; j<col;j++){
      printf("%d ", m[i][j]);
    }
    printf(" \n");
  }
}

int main(){
  int linhas, colunas;
  printf("Digite o tamanho da matriz: ");
  scanf("%d %d",&linhas,&colunas);
  int m[linhas][colunas];
  read_matriz(linhas, colunas,m);
  exibe_matriz(linhas,colunas,m);
  return 0;
}
```

Crie um programa que gere automaticamente uma matriz 10x5 com inteiros entre 0 e 9. O programa deve então criar uma segunda matriz 5x10 que será a transposta da primeira. Ambas as matrizes devem ser exibidas usando a função desenvolvida no exercício anterior.

Instruções adicionais para este exercício:
1 - O tamanho das matrizes pode ser fixado por constantes. Isto pode ser feiro usando a instrução "define" fora da função main. Exemplo: #define M 10
2 - Digitar multiplos valores dá trabalho. Por isso, é útil saber como gerar valores aleatoriamente. A fução rand(), definida na biblioteca stdlib.h, retorna um valor pseudo-aleatório. Se usarmos "int x = rand()%10;", então x receberá um valor aleatório entre 0 e 9. Esta mesma abordagem deverá ser usada na geração da matriz.
3 - Se o programa for executado múltiplas vezes, a sequência de valores aleatórios será sempre igual. Para controlar a geração de valores, é preciso definir uma semente usando o procedimento "srand(seed)". Faça com que o valor de seed seja forneceido pelo usuário, e então gere a matriz aleatória. Observe que com valores diferentes de seed, as matrizes em diferentes execuções terão diverentes valores.


Exemplo de execução:


Digite o valor da seed: 0
A matriz é:                                                                     

3 6 7 5 3
5 6 2 9 1
2 7 0 9 3
6 0 6 2 6
1 8 7 9 2
0 2 3 7 5
9 2 2 8 9  
7 3 6 1 2
9 3 1 9 4
7 8 4 5 0                                                                       

A transposta é:                                                                 

3 5 2 6 1 0 9 7 9 7
6 6 7 0 8 2 2 3 3 8
7 2 0 6 7 3 2 6 1 4  
5 9 9 2 9 7 8 1 9 5
3 1 3 6 2 5 9 2 4 0

```python
%%writefile  matrizesEfuncoes4.c
#include <stdio.h>
#include <stdlib.h>
#define linhas 10
#define colunas 5


void gerando_matrizAletoria(int row,int col,int m[][col]){
  for(int i =0; i<row ; i++){
    for(int j = 0 ;j<col ; j++){
       m[i][j]=rand()%10;
    }
  }
}

void exibe_matriz(int row, int col,int m[][col]){
  printf("A matriz digitada é: \n");
  for(int i =0; i<row ; i++){
    for(int j = 0 ;j<col ; j++){
      printf("%d ",m[i][j]);
    }
    printf("\n");
    }
  }

void gerando_matrizTransposta(int row, int col,int m[][col], int mT[][row]){
  for(int i =0; i<row ; i++){
    for(int j = 0 ;j<col ; j++){
      mT[j][i]=m[i][j];
    }
  }
}

void exibe_matrizT(int row, int col,int mT[][row]){
  printf("A transposta é: \n");
  for(int i =0; i<row ; i++){
    for(int j = 0 ;j<col ; j++){
      printf("%d ",mT[i][j]);
    }
    printf("\n");
    }
  }


int main(){

  printf("Digite o valor da seed: ");
  int seed;
  scanf("%d",&seed);
  srand(seed);

  int m[linhas][colunas]; //delcarando uma matriz 10x5
  int mT[colunas][linhas]; // declarando uma matriz transposta 5x10

  gerando_matrizAletoria(linhas,colunas,m);

  exibe_matriz(linhas,colunas,m);


  gerando_matrizTransposta(linhas,colunas,m,mT);

  exibe_matrizT(colunas ,linhas,mT);

  return 0;
}
```

Soma de matrizes

```python
%%writefile matrizesEfuncoes5.c
#include <stdio.h>

void read_matriz(int row, int col,int m[][col]){
  printf("digite os elementos da matriz :");
  for(int i=0;i<row;i++){
    for(int j = 0; j<col;j++){
      printf("Digite o elemento (%d,%d): \n",i,j);
      scanf("%d",&m[i][j]);}
      }
}


void exibe_matriz(int row, int col, int m[][col]){
  printf(" a  matriz é:\n");
  for(int i=0;i<row;i++){
    for(int j = 0; j<col;j++){
      printf("%d",m[i][j]);}
      }
      printf("\n");
}



void soma_matrizes(int row, int col, int m[][col],int n[][col],int soma[][col]){
  for(int i=0;i<row;i++){
    for(int j = 0; j<col;j++){
      soma[i][j] = m[i][j]+n[i][j];}
      }
}

int main(){

  int l1,l2,c1,c2; // atribuindo linhas(l) e colunas (c)
  printf("Digite o tamanho da matriz 1: ");
  scanf("%d %d",&l1,&c1);
  printf("Digite o tamanho da matriz 2: ");
  scanf("%d %d",&l2,&c2);

  int m1[l1][c1];// construindo as matrizes
  int m2[l2][c2];


  if(l1==l2 && c1==c2){
    int soma[l1][c1];
    read_matriz(l1,c1,m1);
    read_matriz(l2,c2,m2);
    exibe_matriz(l1,c1,m1);
    exibe_matriz(l2,c2,m2);
    soma_matrizes(l1,c1,m1,m2,soma);
    exibe_matriz(l1,c1,soma);
  }
  else{
    printf("As matrizes não podem ser somadas");
  }


  return 0;}
```

soma dos elementos de um vetor:
Faça um programa que some os elementos de um vetor. Dado um vetor v de
tamanho tam, calcular a soma dos elementos da posição 0 até tam-1.

```python
%%writefile vetoresEfuncoes6.c
#include <stdio.h>


int main(){
  int tam;
  printf("Digite o tamanho do vetor: ");
  scanf("%d",&tam);
  int vetor[tam];

  for(int i=0;i<tam;i++){
    printf("Digite o elemento %d do vetor: ",i);
    scanf("%d",&vetor[i]);
  }

  int soma = 0;

  for(int i = 0; i<tam; i++){
    soma  = soma +vetor[i];
  }

  printf("A soma dos elementos do vetor é: %d",soma);


  return 0;
}
```
