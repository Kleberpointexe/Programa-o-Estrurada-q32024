```python
%%writefile operacoesEponteiros.c
#include <stdio.h>

int x,y,r;
int *px,*py,*pr;

int main(){
   printf("Digite dois valores inteiros:");
    scanf("%d %d",&x,&y);
    px  =&x;
    py = &y;
    pr = &r;
    *pr = *px+*py;
    printf("A soma é : %d\n",*pr);
    *pr = *px*(*py);
    printf("A mutilicação é: %d\n",*pr);
    *pr = (*px)/(*py);
    printf("A divisão é: %d\n",*pr);
    *pr = *px%*py;
    printf("O resto é: %d\n",*pr);
  return 0 ;
}
```

```python
%%writefile ponteirosEvetores.c
#include <stdio.h>
#include <stdlib.h>

int main() {
    int n; // Tamanho do vetor
    int *v; // Ponteiro para o vetor
    int *p; // Ponteiro auxiliar

    // Solicita o número de elementos do vetor
    printf("Digite o número de elementos do vetor: ");
    scanf("%d", &n);

    // Aloca dinamicamente o vetor
    v = (int *)malloc(n * sizeof(int));
    if (v == NULL) {
        printf("Erro de alocação de memória.\n");
        return 1;
    }

    p = v; // Inicializa o ponteiro auxiliar

    // Lê os elementos do vetor
    printf("Digite os elementos do vetor:\n");
    for (int i = 0; i < n; i++) {
        scanf("%d", (p + i));
    }

    // Exibe os elementos do vetor
    for (int i = 0; i < n; i++) {
        printf("O elemento na posição %d é %d.\n", i, *(p + i));
    }

    // Libera a memória alocada
    free(v);

    return 0;
}
```

```python
%%writefile PonteirosEvetorDeString.c
#include <stdio.h>
#include <string.h>

int main() {
    char texto[100]; // Vetor para armazenar a string
    char *str = texto; // Ponteiro para o vetor de caracteres
    int comprimento_p = 0; // Comprimento calculado manualmente

    // Solicita ao usuário que entre com a string
    printf("Entre uma string: ");
    fgets(texto, sizeof(texto), stdin);

    // Remove o caractere de nova linha (\n) inserido pelo fgets, se presente
    size_t len = strlen(texto);
    if (texto[len - 1] == '\n') {
        texto[len - 1] = '\0';
    }

    // Calcula o comprimento da string usando o ponteiro
    while (*(str + comprimento_p) != '\0') {
        comprimento_p++;
    }

    // Calcula o comprimento da string usando strlen
    int comprimento_strlen = strlen(texto);

    // Exibe os resultados
    printf("O comprimento usando ponteiro é: %d\n", comprimento_p);
    printf("O comprimento usando strlen é: %d\n", comprimento_strlen);

    return 0;
}
```

```python
%%writefile PonteirosEtamanhoDoVetor.c
#include <stdio.h>
#include <stdlib.h>

int main() {
    int N; // Tamanho do vetor
    int *vetor; // Ponteiro para o vetor
    int posicao; // Posição a ser consultada

    // Solicita o tamanho do vetor
    printf("Tamanho do vetor: ");
    scanf("%d", &N);

    // Aloca dinamicamente o vetor
    vetor = (int *)malloc(N * sizeof(int));
    if (vetor == NULL) {
        printf("Erro de alocação de memória.\n");
        return 1;
    }

    // Lê os elementos do vetor
    printf("Vetor:\n");
    for (int i = 0; i < N; i++) {
        scanf("%d", (vetor + i));
    }

    // Solicita a posição a ser consultada
    printf("Elemento: ");
    scanf("%d", &posicao);

    // Verifica se a posição é válida
    if (posicao >= 0 && posicao < N) {
        printf("O elemento %d vale: %d\n", posicao, *(vetor + posicao));
    } else {
        printf("Posição inválida.\n");
    }

    // Libera a memória alocada
    free(vetor);

    return 0;
}
```

```python
%%writefile VetorEpassagemPorParametros.c
#include <stdio.h>
#include <stdlib.h>

// Função para leitura do vetor
void read_vec(int *p, int N) {
    printf("Digite os elementos do vetor:\n");
    for (int i = 0; i < N; i++) {
        scanf("%d", (p + i));
    }
}

// Função para exibição do vetor
void show_vec(int *p, int N) {
    printf("O vetor é: ");
    for (int i = 0; i < N; i++) {
        printf("%d ", *(p + i));
    }
    printf("\n");
}

// Função para inverter o vetor
void inv_vec(int *p, int N) {
    int temp;
    for (int i = 0; i < N / 2; i++) {
        temp = *(p + i);
        *(p + i) = *(p + (N - 1 - i));
        *(p + (N - 1 - i)) = temp;
    }
}

int main() {
    int N; // Tamanho do vetor
    int *vetor; // Ponteiro para o vetor

    // Solicita o tamanho do vetor
    printf("Digite o tamanho do vetor: ");
    scanf("%d", &N);

    // Aloca dinamicamente o vetor
    vetor = (int *)malloc(N * sizeof(int));
    if (vetor == NULL) {
        printf("Erro de alocação de memória.\n");
        return 1;
    }

    // Chama as funções para leitura, exibição e inversão do vetor
    read_vec(vetor, N);
    show_vec(vetor, N);
    inv_vec(vetor, N);
    show_vec(vetor, N);

    // Libera a memória alocada
    free(vetor);

    return 0;
}
```
