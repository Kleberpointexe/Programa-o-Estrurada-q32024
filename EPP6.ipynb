```python
%%writefile Estrutura_simples.c
#include <stdio.h>
#include <string.h>

//delcarando uma estrutura
struct user {
    char nome[50];
    int idade;
    float altura;
    float peso;
};

//construindo função para mostrar estrutura

void showStruct(struct user pessoa){
    printf("O nome é:%s \n ",pessoa.nome);
    printf("A idade é:%d \n",pessoa.idade);
    printf("A altura é:%.2f \n",pessoa.altura);
    printf("O peso é:%.2f \n",pessoa.peso);
};

int main(){

    // instanciando pessoa01
    struct user pessoa01,pessoa02;
    strcpy(pessoa01.nome,"João Silva ");
    pessoa01.idade = 30;
    pessoa01.altura = 1.65;
    pessoa01.peso = 75.50;

    //instanciando pessoa02 pelo usuário
    printf("Digite o nome:\n");
    fgets(pessoa02.nome, 50 ,stdin);
    printf("Digite a idade:\n");
    scanf("%d", &pessoa02.idade);getchar();
    printf("Digite a altura:\n");
    scanf("%f",&pessoa02.altura);getchar();
    printf("Digite o peso:\n");
    scanf("%f",&pessoa02.peso);getchar();

    //mostrando os registros das pessoas
    showStruct(pessoa01);
    showStruct(pessoa02);

    return 0;
}
```

Crie um programa que use uma estrutura para armazenar informações de alunos de uma turma. A estrutura deve conter nome, frequência e nota final. O programa deve avaliar quantos alunos obtiveram nota acima de 5, quantos foram reprovados por falta (frequência abaixo de 75%) e a média da turma.
Utilize uma estrutura: struct turma{ ... };


Exemplo de execução:
2                                                                               
Ana                                                                               
100                                                                             
10                                                                              
Maria                                                                               
90                                                                               
8                                                                               
2                                                                              
0                                                                             
9.00  





Obs: Não utilize o typedef no laboratório.

```python
%%writefile Estrutura_e_condicionais.c
#include <stdio.h>
#include <string.h>

// Estrutura para armazenar as informações dos alunos
struct turma {
    char nome[100];  // Nome do aluno
    int frequencia;  // Frequência do aluno (em %)
    float nota;      // Nota final do aluno
};

int main() {
    int n, acimaDeCinco = 0, reprovadosPorFalta = 0;
    float somaNotas = 0.0;

    // Leitura do número de alunos
    scanf("%d", &n);

    struct turma alunos[n]; // Vetor de estruturas

    // Leitura dos dados de cada aluno
    for (int i = 0; i < n; i++) {
        scanf("%s", alunos[i].nome);       // Nome do aluno
        scanf("%d", &alunos[i].frequencia); // Frequência do aluno
        scanf("%f", &alunos[i].nota);      // Nota do aluno

        // Verifica se o aluno obteve nota acima de 5
        if (alunos[i].nota > 5) {
            acimaDeCinco++;
        }

        // Verifica se o aluno foi reprovado por falta (frequência abaixo de 75%)
        if (alunos[i].frequencia < 75) {
            reprovadosPorFalta++;
        }

        // Soma as notas para calcular a média da turma
        somaNotas += alunos[i].nota;
    }

    // Calcula a média das notas
    float mediaTurma = somaNotas / n;

    // Exibe os resultados
    printf("%d\n", acimaDeCinco);            // Número de alunos com nota acima de 5
    printf("%d\n", reprovadosPorFalta);      // Número de alunos reprovados por falta
    printf("%.2f\n", mediaTurma);            // Média da turma com 2 casas decimais

    return 0;
}
```

Crie um programa que armazene o resultado de uma  operação aritmética em uma união "value". O programa principal deve pedir que sejam digitados os operandos inteiros x e y. A opção da operação também deve ser digitada, sendo 1-Adição, 2-Subtração, 3-Multiplicação e 4-Divisão. Para adição, subtração e multiplicação, o resultado é armazenado como inteiro na união. Para a divisão, o resultado é armazenado como um vetor de 2 posições, sendo 1 inteiro para o quociente e 1 inteiro para o resto. Ainda para a divisão, caso o divisor for 0, o resultado é armazenado como a string "Erro". O resultado deve então ser exibido em uma mensagem.
Utilize a união: union value{ ... };
Exemplo de execução:

Digite os operandos x e y                                                       
1                                                                               
2                                                                               
Digite a operação: 1 - Adição, 2 - Subtração, 3 - Multiplicação , 4 - Divisão   
3                                                                               
Resultado: 2  

```python
%%writefile union.c



#include <stdio.h>
#include <string.h>

union value {
    int adição, subtração, multiplicação;
    int divisão[2];
    char erro[6];
};

int main() {
    int x, y;
    union value resultado;

    printf("Digite os operandos x e y: \n");
    scanf("%d%d", &x, &y);

    printf("Digite a operação: 1 - Adição, 2 - Subtração, 3 - Multiplicação, 4 - Divisão \n");

    int opcao = 0;
    scanf("%d", &opcao);

    switch (opcao) {
        case 1:
            resultado.adição = x + y;
            printf("Resultado:%d \n", resultado.adição);
            break;

        case 2:
            resultado.subtração = x - y;
            printf("Resultado:%d \n", resultado.subtração);
            break;

        case 3:
            resultado.multiplicação = x * y;
            printf("Resultado:%d \n", resultado.multiplicação);
            break;

        case 4:
            if (y == 0) {
                strcpy(resultado.erro, "Erro");
                printf("Resultado: %s\n", resultado.erro);
            } else {
                resultado.divisão[0] = x / y;
                resultado.divisão[1] = x % y;
                printf("Resultado: quociente:%d, resto:%d \n", resultado.divisão[0], resultado.divisão[1]);
            }
            break;

        default:
            printf("Opção inválida!\n");
            break;
    }

    return 0;
}
```
